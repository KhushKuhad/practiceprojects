```python
bios_new = bios.copy()
bios_new['first_name'] = bios_new['name'].str.split(' ').str[0]
bios_new
```

```python
bios_new['born_datetime'] = pd.to_datetime(bios_new['born_date']) #converting to datetime type
bios_new.info()
```

```python
bios_new['born_year'] = bios_new['born_datetime'].dt.year
bios_new[['name','born_year']] #using extra datetime features to extract specicific data like only year 
```

### Lambda funtion for pandas advanced functions and .apply()

```python
bios_new['weight_class'] = bios_new['weight_kg'].apply(lambda x: 'lightweight' if x < 70 else('middleweight' if x < 90 else 'Heavyweight') )
#CAN ALSO MAKE AND USE CUSTOM FUNCTIONS INSIDE APPLY
```

```python
bios_new
```

```python
nocs = pd.read_csv('noc_regions.csv')
bios_new = pd.merge(bios_new,nocs,left_on='born_country',right_on='NOC',how='left')
bios_new.rename(columns={'region':'born_country_fullname'},inplace=True)
bios_new #joining 2 tables using left join, matches data from the nocs data and fits in the old table
```

```python
usa = bios[bios["born_country"] == 'USA'].copy()
gbr = bios_new[bios_new["born_country"] == 'GBR'].copy()
usa.head()
```

```python
gbr.head()
```

```python
concated_df = pd.concat([usa,gbr]) #CONCATENATION OF TWO TABLES
concated_df
```

### Handling Null values

```python
ir_data = data.head(10).copy()
ir_data
```

```python
ir_data.loc[[0,7],'petal_width'] = np.nan
ir_data
```

```python
#fill missing values using mean
ir_data.fillna(ir_data['petal_width'].mean())
```

```python
#we can drop the rows
ir_data.dropna(subset=['petal_width']) #we use subset so it only drops if "petal_width" is Nan
```

```python
#filter data to get nan
ir_data[ir_data['petal_width'].isna()]
```

```python
#filter data to avoid nans
ir_data[ir_data['petal_width'].notna()]
```
